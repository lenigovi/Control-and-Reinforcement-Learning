### Aufgabe 1

Gegeben ist das System
$$ \dot{x} = \begin{bmatrix}a&1\\2&b\end{bmatrix}x  \ \ \ \ \ \ \ \ \ y = \begin{bmatrix}0&1\end{bmatrix}x \ \ \ \ \ \ \ \ mit \ a,b \in R $$




Die Zustände des Systems sollen mit einem Luenberger-Beobachter mit der Matrix $ L = \begin{bmatrix}l_1&l_2\end{bmatrix}^T$ geschätzt werden. Die Pole des Beobachters sollen dabei bei $-2$ liegen. Geben Sie den Wert für $l_1$ in Abhängigkeit von $a$ und $b$ an.


---

Der Beobachter hat die Form:

$ \dot{\hat{x}} = A\hat{x} + L(y - \hat{y}) = A\hat{x} + L(Cx - C\hat{x}) $


wobei $ A = \left[\begin{array}{cc}a & 1 \\2 & b\end{array}\right]$ und $C = \left[\begin{array}{ll}0 & 1\end{array}\right]$ und die Matrix L ist gegeben durch $L = \left[\begin{array}{c}l_1 \\ l_2\end{array}\right]$.

Die Fehlerdynamik des Beobachtersystems ist:

$$
\dot{e} = \dot{x} - \dot{\hat{x}} = Ax - (A\hat{x} + L(Cx - C\hat{x})) = (A - LC)e
$$


wobei $e = x - \hat{x}$ der Zustandsfehler ist. Die Matrix $A - LC$ soll die Pole bei $-2$ haben, also:

$$
A - LC = \left[\begin{array}{cc}
a & 1 \\
2 & b
\end{array}\right] - \left[\begin{array}{c}
l_1 \\
l_2
\end{array}\right] \left[\begin{array}{cc}
0 & 1
\end{array}\right] = \left[\begin{array}{cc}
a & 1 - l_1 \\
2 & b - l_2
\end{array}\right]
$$


Die $ 2 \times 2 $ Matrix soll die Wurzeln $\lambda = -2 $ haben. Für eine $ 2 \times 2 $ Matrix ist die charakteristische Gleichung:

$$
\lambda^2 - \text{Tr}(A - LC) \lambda + \det(A - LC) = 0
$$

$ \text{Tr}(A - LC) = a + b - l_2 $
$ \det(A - LC) = a(b - l_2) - (1 - l_1)2 $

- $ \text{Tr}(A - LC) = \lambda + \lambda = -4 $
- $ \det(A - LC) = \lambda \times \lambda = 4 $


Dann haben wir:

$ a + b - l_2 = -4 \Rightarrow l_2 = a + b + 4 $

und wenn wir $l_2$ in die Determinantenbedingung einsetzen:


$ a(b - (a + b + 4)) - 2(1 - l_1) = 4 \\
a(-a - 4) - 2 + 2l_1 = 4 \\
2l_1 = 4 + 2 + a^2 + 4a $

$l_1 = \frac{a^2 + 4a + 6}{2}$

Gegeben ist ein Signal Input Signal Output (SISO) System

$$ ||G(j\omega)|| = sup|G(j\omega)| $$ 
(maximale Verstärkung)



```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameter range
theta_range = np.linspace(0, 2*np.pi, 100)

# Initialize arrays to store J values
J_values = []

# Calculate J for each theta
for theta in theta_range:
    # Define Q matrix
    Q = np.array([[np.cos(theta), 0],
                  [0, np.sin(theta)]])
    # Define R matrix
    R = np.array([[1]])
    # Calculate J (for simplicity, assuming a constant value)
    J = np.sum(Q) + np.sum(R)
    J_values.append(J)

# Plotting
plt.plot(theta_range, J_values)
plt.xlabel('Theta')
plt.ylabel('Performance Index (J)')
plt.title('Performance Index vs. Parameter Theta')
plt.grid(True)
plt.show()
```
